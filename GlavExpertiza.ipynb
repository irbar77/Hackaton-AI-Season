```python
name_text = "РЕКОНСТРУКЦИЯ МОСТА ЧЕРЕЗ РЕКУ"
```

```python
# Импортируем библиотеку для NLP
import nltk

# Токенизируем строку name_text
name_text_tokens = nltk.word_tokenize(name_text)

# Токенизируем текст content
content_tokens = nltk.word_tokenize(content)

# Приводим name_text_tokens и content_tokens к нижнему регистру
name_text_tokens = [token.lower() for token in name_text_tokens]
content_tokens = [token.lower() for token in content_tokens]


# Создаем переменную для хранения счетчика
count = 0

# Цикл, который проходит по всем словам name_text_tokens
for token in name_text_tokens:
    # Проверяем есть ли такое слово в тексте content_tokens
    if token in content_tokens:
        # Если да, увеличиваем счетчик
        count += 1

# Выводим результат
print('Текст name_text встретился в тексте content', count, 'раз.')
```

```python
# Импортируем нужные библиотеки
import spacy
from spacy import displacy

# Загружаем модель для использования
nlp = spacy.load('en_core_web_sm')

# Создаем документ для анализа
doc = nlp(name_text + content)

# Выполняем Named Entity Recognition
for ent in doc.ents:
    print(ent.text, ent.label_)

# Визуализируем извлеченные сущности
displacy.render(doc, style='ent', jupyter=True)
```

```python
name_text12 = "РЕКОНСТРУКЦИЯ МОСТА ЧЕРЕЗ РЕКУ ФАРС НА КМ 72+711 АВТОМОБИЛЬНОЙ ДОРОГИ Р-217 «КАВКАЗ» АВТОМОБИЛЬНАЯ ДОРОГА М-4 «ДОН» - ВЛАДИКАВКАЗ – ГРОЗНЫЙ – МАХАЧКАЛА - ГРАНИЦА С АЗЕРБАЙДЖАНСКОЙ РЕСПУБЛИКОЙ. ПОДЪЕЗД К Г. МАЙКОП, КРАСНОДАРСКИЙ КРАЙ»"
name_text13 = "РЕКОНСТРУКЦИЯ МОСТА ЧЕРЕЗ РЕКУ ФАРС НА КМ 72+711 АВТОМОБИЛЬНОЙ ДОРОГИ Р-217 «КАВКАЗ»  «КАВКАЗ»  АВТОМОБИЛЬНАЯ ДОРОГА М-4 «ДОН» - ВЛАДИКАВКАЗ – ГРОЗНЫЙ – МАХАЧКАЛА - ГРАНИЦА С АЗЕРБАЙДЖАНСКОЙ РЕСПУБЛИКОЙ. ПОДЪЕЗД К Г. МАЙКОП, КРАСНОДАРСКИЙ КРАЙ»"
w1 = nlp(name_text12)
w2 = nlp(name_text13)

similarity = w1.vector.dot(w2.vector) / (w1.vector_norm * w2.vector_norm)
print(similarity)
```

```python
import spacy
from spacy.tokens import Span

nlp = spacy.load("en_core_web_sm")

name_text12 = "РЕКОНСТРУКЦИЯ МОСТА ЧЕРЕЗ РЕКУ ФАРС НА КМ 72+711 АВТОМОБИЛЬНОЙ ДОРОГИ Р-217 «КАВКАЗ» АВТОМОБИЛЬНАЯ ДОРОГА М-4 «ДОН» - ВЛАДИКАВКАЗ – ГРОЗНЫЙ – МАХАЧКАЛА - ГРАНИЦА С АЗЕРБАЙДЖАНСКОЙ РЕСПУБЛИКОЙ. ПОДЪЕЗД К Г. МАЙКОП, КРАСНОДАРСКИЙ КРАЙ»"
name_text13 = "РЕКОНСТРУКЦИЯ МОСТА ЧЕРЕЗ РЕКУ ФАРС НА КМ 72+711 АВТОМОБИЛЬНОЙ ДОРОГИ Р-217 «КАВКАЗ»  «КАВКАЗ»  АВТОМОБИЛЬНАЯ ДОРОГА М-4 «ДОН» - ВЛАДИКАВКАЗ – ГРОЗНЫЙ – МАХАЧКАЛА - ГРАНИЦА С АЗЕРБАЙДЖАНСКОЙ РЕСПУБЛИКОЙ. ПОДЪЕЗД К Г. МАЙКОП, КРАСНОДАРСКИЙ КРАЙ»"

doc1 = nlp(name_text12)
doc2 = nlp(name_text13)

# Create a list of tokens to compare
tokens1 = [token.text for token in doc1]
tokens2 = [token.text for token in doc2]

# Get the differences between the tokens
differences = [token for token in tokens2 if token not in tokens1]

# Create a Span for each token
spans = [Span(doc2, tokens2.index(token), tokens2.index(token) + 1) for token in differences]

# Visualize the differences
displacy.render(doc2, style="ent", jupyter=True, options={"ents": ["TOKEN" for token in differences]}, manual=spans)
```
